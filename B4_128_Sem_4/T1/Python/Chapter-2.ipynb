### Simple Area Chart

```python
import matplotlib.pyplot as plt
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='wheat',alpha=0.8)
plt.plot(x,y,color='brown',alpha=0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expenses,color='red')
plt.fill_between(time,income,expenses,where=(income>expenses),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expenses,where=(income<=expenses),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel('time')
plt.ylabel('Income/Expenses Compairision')
plt.legend()
plt.show()
```

### Stacked Area Chart

```python
import matplotlib.pyplot as plt
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
plt.stackplot(x,y1,y2,y3,labels=['A','B','C'],colors=['black','yellow','red'])
plt.legend(loc='upper left')
plt.show()
```

### Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt
x1=5*np.random.rand(40)
x2=5*np.random.rand(40)+25
x3=25*np.random.rand(20)
x=np.concatenate((x1,x2,x3))
y1=5*np.random.rand(40)
y2=5*np.random.rand(40)+25
y3=25*np.random.rand(20)
color_array=['b']*40+['g']*40+['r']*20
# print(color_array)
y=np.concatenate((y1,y2,y3))
plt.scatter(x,y,s=[100],marker='^',color=color_array)
plt.show()
```

### Waffle Chart 

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
data={'phone':['Xiomi','Samsung','Apple','Nokia','Realme'],'stock':[45,12,8,5,50]}
df=pd.DataFrame(data)
fig=plt.figure(FigureClass=Waffle,rows=10,values=df.stock,labels=list(df.phone))
```

### Word Cloud

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
alice_novel=open('alice_in_wonderland.txt')
alice_novel=alice_novel.read()
stopwords=set(STOPWORDS)
stopwords.add('said')
# print(stopwords)
# print(len(stopwords))
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
alice_wc.generate(alice_novel)
plt.figure(figsize=[15,15])
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
G=nx.DiGraph()
l=[[0,1,0,1],[0,0,0,1],[1,0,0,0],[1,0,1,0]]
G.add_nodes_from([0,1,2,3])
for i in range(len(l)):
    for j in range(len(l[i])):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw_networkx(G,node_size=850,node_color='red',width=2,edge_color='blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
l={'A':[1,0,0,1,0,1,1],'B':[0,1,1,1,1,1,0],'C':[1,1,1,0,1,0,1],'D':[1,0,0,0,1,0,1],'E':[1,0,0,0,1,1,0],'F':[1,1,1,1,1,0,1],'G':[0,1,1,0,0,0,1]}
G=nx.DiGraph()
G.add_nodes_from(l.keys())
key_list=list(l.keys())
for key,value in l.items():
    for i in range(len(value)):
        if value[i]==1:
            G.add_edge(key,key_list[i])
nx.draw_networkx(G,node_size=850,node_color='red',width=2,edge_color='blue')
plt.show()
```
