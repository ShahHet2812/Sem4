```python
m=pd.read_csv('auto-mpg.csv')
def find_outliner(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-1.5*IQR
    high_val=q3+1.5*IQR
    print("low val :",low_val,"high_val :",high_val)
    ds=ds.loc[(ds[col]<low_val) | (ds[col]>high_val)]
    return ds

print(find_outliner(m,'mpg'))

```

```python
m=pd.read_csv('auto-mpg.csv')
def remove_outliner(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-1.5*IQR
    high_val=q3+1.5*IQR
    print("low val :",low_val,"high_val :",high_val)
    ds=ds.loc[(ds[col]>=low_val) & (ds[col]<=high_val)]
    return ds

print(remove_outliner(m,'mpg'))

```
