##  Politifacts.com/factchecks

```python
import requests
from bs4 import BeautifulSoup
website='https://politifact.com/factchecks/'
result=requests.get(website)
content=result.text
```

```python
fact=soup.find('li',class_='o-listicle__item')
fact
```

```python
fact=soup.find_all('li',class_='o-listicle__item')
fact
```

```python
statement=[]
for i in fact:
    state=i.find('div',class_='m-statement__quote').text.strip()
    statement.append(state)
statement
```

```python
source=[]
for i in fact:
    src=i.find('div',class_='m-statement__meta').find('a').text.strip()
    source.append(src)
print(source)
```

```python
dates=[]
for i in fact:
    date=i.find('footer',class_='m-statement__footer').text.split('•')[1].strip()
    dates.append(date)
print(dates)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Fact':statement,'Source':source,'Date':dates})
df
```
