```python
# a43d48e2c6398946dd5515d9a1e4a208
import requests,json
import pandas as pd
def get_coordinates():
    api_key="a43d48e2c6398946dd5515d9a1e4a208"
    city=input("Enter City Name:")
    request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response=requests.get(request_url)
    coordinates=response.json()
    latitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    return(latitude,longitude)
def get_weather(lat,lon):
    api_key="a43d48e2c6398946dd5515d9a1e4a208"
    request_url=f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    weather=response.json()
#    print(json.dumps(weather,indent=5))
    print("Description:"+weather['weather'][0]['description'])
    print("Temperature(Degree Celcius):"+str(weather['main']['temp']-273.15))
    print("Pressure:"+str(weather['main']['pressure']))
    print("Humidity:"+str(weather['main']['humidity']))
    print("Wind Speed:"+str(weather['wind']['speed']))
    print("Visibility:"+str(weather['visibility']))
def aqi(lat,lon):
    api_key="a43d48e2c6398946dd5515d9a1e4a208"
    request_url=f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    air=response.json()
#   print(air)
    aqi=air['list'][0]['main']['aqi']
    dict={1:"Good",2:"Fair",3:"Moderate",4:"Poor",5:"Very Poor"}
    print("Air Quality:"+dict[aqi])
def five_day(lat,lon):
    api_key="a43d48e2c6398946dd5515d9a1e4a208"
    request_url=f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    forecast=response.json()
#     print(json.dumps(forecast,indent=5))
    return forecast
lat,lon=get_coordinates()
get_weather(lat,lon)
aqi(lat,lon)
forecast=five_day(lat,lon)
Data={"date_time":[],"temp":[],"pressure":[],"humidity":[],"weather":[]}
for i in forecast['list']:
    Data['date_time'].append(i['dt_txt'])
    Data['temp'].append(i['main']['temp']-273)
    Data['pressure'].append(i['main']['pressure'])
    Data['humidity'].append(i['main']['humidity'])
    Data['weather'].append(i['weather'][0]['description'])
df=pd.DataFrame(Data)
df
```
