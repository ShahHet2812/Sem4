## Regression-Model Training and Evaluation

#### Linear Regression
Statistical Method used to model the relationship between a dependent variable and one or more independent variable
The goal of Linear Regression is to find the best fitting line that describes relationship between the variables
y--> Dependent Variable
x--> Independent Variable
y=mx+c(y=coefficient,c=intercept)
y=B0+B1x+E(B0=intercept,B1=Slope of line,E=Error Term(residual))

```python
import pandas as pd
data=pd.read_csv('data.csv')
data
```

```python
data.dropna(inplace=True)
data
```

```python
x=data.iloc[:,0:1]
y=data.iloc[:,1]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({'Actual':y,'Predicted':y_pred})
print(diff)
```

```python
print("Coefficient:",lr.coef_)
print("Intercept:",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y,y_pred))
print("MSE:",metrics.mean_squared_error(y,y_pred))
print("R2 score:",metrics.r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,y_pred,c='r')
plt.show()
```
