## Regression-Model Training and Evaluation

#### Linear Regression
Statistical Method used to model the relationship between a dependent variable and one or more independent variable
The goal of Linear Regression is to find the best fitting line that describes relationship between the variables
y--> Dependent Variable
x--> Independent Variable
y=mx+c(y=coefficient,c=intercept)
y=B0+B1x+E(B0=intercept,B1=Slope of line,E=Error Term(residual))

```python
import pandas as pd
data=pd.read_csv('data.csv')
data
```

```python
data.dropna(inplace=True)
data
```

```python
x=data.iloc[:,0:1]
y=data.iloc[:,1]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({'Actual':y,'Predicted':y_pred})
print(diff)
```

```python
print("Coefficient:",lr.coef_)
print("Intercept:",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y,y_pred))
print("MSE:",metrics.mean_squared_error(y,y_pred))
print("R2 score:",metrics.r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c='b',label='Actual')
plt.plot(x,y_pred,c='r',label='Predicted')
plt.legend()
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('Book1.csv')
data
```

```python
cgpa=pd.DataFrame(data.cgpa)
package=pd.Series(data.package)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(cgpa,package,test_size=0.2,random_state=16)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y_test,y_pred))
print("MSE:",metrics.mean_squared_error(y_test,y_pred))
print("R2 score:",metrics.r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.cgpa,data.package,c='b',label='Actual Data')
plt.plot(x_test,y_pred,c='r',label='Predicted Data')
plt.legend()
plt.show()
```

```python
l=[]
for i in range(0,44):
    x_train,x_test,y_train,y_test=train_test_split(cgpa,package,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2=metrics.r2_score(y_test,y_pred)
    l.append(r2)
print(l)
plt.plot(range(44),l,marker='o')
plt.grid()
```
